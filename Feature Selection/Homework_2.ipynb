### (b) Fit a ridge regression model on the training set, with λ chosen by cross- validation. Report the test error obtained.

```r
install.packages('glmnet')
library(glmnet)
```

Choosing the lambda

```r

matrix_train <-  model.matrix(Apps ~ . -1, data = Train)
matrix_test <-  model.matrix(Apps ~ . -1, data = Test)

grid <-  10 ^ seq(4, -2, length = 100)

ridge_model <-  cv.glmnet(matrix_train, Train[, "Apps"], 
                        alpha = 0, lambda = grid, thresh = 1e-12)
best_lambda <-  ridge_model$lambda.min
best_lambda
```

```r
pred_ridg_model <-  predict(ridge_model, newx = matrix_test, s = best_lambda)
mean((Test[, "Apps"] - pred_ridg_model)^2)
```

```r
pred_lasso_model <-  predict(lasso_model, newx = matrix_test, s = best_lambda)
mean((Test[, "Apps"] - pred_lasso_model)^2)
```

### (g) Comment more generally on the results obtained. How accurately can we predict the number of college applications received? Is there much difference among the test errors resulting from these five approaches?

```r
test.avg <-  mean(Test[, "Apps"])

lm.test.r2 <-  1 - mean((Test[, "Apps"] - pred_line_model)^2) /
  mean((Test[, "Apps"] - test.avg)^2)

ridge.test.r2 <-  1 - mean((Test[, "Apps"] - pred_ridg_model)^2)/
  mean((Test[, "Apps"] - test.avg)^2)

lasso.test.r2 <-  1 - mean((Test[, "Apps"] - pred_lasso_model)^2)/
  mean((Test[, "Apps"] - test.avg)^2)

barplot(c(lm.test.r2, ridge.test.r2, lasso.test.r2),
        col = "gray", names.arg = c("Linear", "Ridge", "Lasso"),
        main = "Test R-squared", las=2, ylim=c(0:1))
```

### Forward Stepwise selection


Stepwise forward selection for variable selection.

```r
regfit.fwd=regsubsets(V86~.,data=train,nvmax =85,method ="forward") 
summary(regfit.fwd)
names(summary(regfit.fwd))
```

Here we are taking the CP as the parameter as they indicate smaller amounts of unexplained error.

```r
plot(regfit.fwd,scale="Cp")
coef(regfit.fwd,which.min(summary(regfit.fwd)$cp))
```

We can see that 23 varibales have the least error rate and the coefficient of each is mentioned above.

```r
coef(regfit.fwd,which.min(summary(regfit.fwd)$cp))
```

### Backward Stepwise Selection

```r
regfit.bwd=regsubsets(V86~.,data=train,nvmax =85,method ="backward")
summary(regfit.bwd)
```

```r
plot(regfit.bwd,scale="Cp")
coef(regfit.bwd,which.min(summary(regfit.bwd)$cp))
```

Here also we use the CP varibles for the selection, as we can see in the backward the selection the minimum of CP lower that that of the forward's CP.

We can also assess fit by analysing R2 and adjusted r-squared. These are often used in along with Cp to assess model fit. Since AIC is almost similar to Cp we are not considering it here.

### Validation

```r
test['V86'] = data.frame(read.table('tictgts2000.txt'))
```

```r
set.seed(2312)
regfit.best=regsubsets(V86~.,data=train,nvmax =85, method='forward') 
train.mat=model.matrix(V86~.,data=test)
val.errors =rep(NA ,85)
for(i in 1:85){
    coefi=coef(regfit.best,id=i) 
    pred=train.mat[,names(coefi)]%*% coefi 
    val.errors[i]= mean((test$V86-pred)^2)
} 
val.errors
which.min(val.errors)
coef(regfit.best,27)
```

We can see as the number of varible increases the train error getting lower till a particular point then it became stable. Here at 27 it gets the minimum which measn we hve the lowest error rate with 27 varibles. the coeffiient of each 27 varibles plotted above.

```r
set.seed(25562)
regfit.best=regsubsets(V86~.,data=train,nvmax =85, method='backward') 
train.mat=model.matrix(V86~.,data=test)
val.errors =rep(NA ,85)
for(i in 1:85){
    coefi=coef(regfit.best,id=i) 
    pred=train.mat[,names(coefi)]%*% coefi 
    val.errors[i]= mean((test$V86-pred)^2)
} 
val.errors
which.min(val.errors)
coef(regfit.best,38)
```

Although they didn't choose the same models, in this case they both discovered the model that reduced the Cp globally.

Forward selection selected model with 27 varibales and backward selection selected model with 38 varibles.

When there are many variables to consider,forward stepwise selection might be employed instead of backward. This is so that unlike backward selection, which must take into account the entire model, forward selection begins with a null model (with no predictors) and adds variables one at a time (which includes all the predictors).
