```r
# Trianing model.

model <- lm(mpg ~ ., data = Auto)

# Summary of model.
summary(model)
```

```r
# a. Which predictors appear to have a significant relationship to the response. 

Displacement, weight, year, and origin all have statistically significant relationships with the respose as the Pr(>|t|)(p-value) of these variables are less than significance value of  = .05.
```

```r
# b. What does the coefficient variable for year suggest?

The coeffcient of the varible year is 0.750773, which means that increase of mpg for every additional one additional year.
If the year increases by 1 unit, the average mgp increases by 0.750773 units.
```

```r
# c. Use the * and : symbols to fit some models with interactions. Are there any interactions that are significant? (You do not need to select all interactions)

# Using *

lm_fit_interaction1 <-  lm(mpg ~ displacement * horsepower, data = Auto)
summary(lm_fit_interaction1)


# There is statistical significant between 'displacement' and 'horsepower'.


lm_fit_interaction2 <-  lm(mpg ~ cylinders * origin , data = Auto)
summary(lm_fit_interaction2)

# There is no statistical significant between 'cylinders' and 'origin'.
```

```r
# Using :

lm_fit_interaction1 <-  lm(mpg ~ cylinders : displacement , data = Auto)
summary(lm_fit_interaction1)


# There is statistical significant between 'cylinders' and 'displacement'.


lm_fit_interaction2 <-  lm(mpg ~ horsepower : origin , data = Auto)
summary(lm_fit_interaction2)

# There is no statistical significant between 'horsepower' and 'origin'.
```
